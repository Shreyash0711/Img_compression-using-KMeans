```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from skimage import io
```

```python
url = 'https://www.gstatic.com/webp/gallery/4.jpg'
img_original = io.imread(url)
img = np.array(img_original, dtype=float) / 255
h, w, c = img.shape
n_pixels = h * w
image = img.reshape([n_pixels, 3])
```

```python




def compress_with_kmeans(k):
    KM = KMeans(n_clusters=k, random_state=123)
    KM.fit(image)

    
    compressed = np.zeros([n_pixels, 3])
    for i in range(k):
        compressed[KM.labels_ == i] = KM.cluster_centers_[i]

    
    compressed_img = compressed.reshape([h, w, 3])
    compressed_img_scaled = np.array(compressed_img * 255, dtype=np.uint8)

    
    original_bits = n_pixels * 3 * 8
    cluster_centers_bits = k * 3 * 32
    label_bits = n_pixels * np.ceil(np.log2(k))
    compressed_bits = int(cluster_centers_bits + label_bits)

    
    original_kb = original_bits / 8 / 1024
    compressed_kb = compressed_bits / 8 / 1024
    compression_ratio = original_bits / compressed_bits

  
    print(f"--- K = {k} ---")
    print(f"Cluster Count: {k}")
    print(f"Image Size: {h} x {w} = {n_pixels} pixels")
    print(f"Original Size: {original_kb:.2f} KB")
    print(f"Compressed Size: {compressed_kb:.2f} KB")
    print(f"Compression Ratio: {compression_ratio:.2f}\n")

    
    plt.imshow(compressed_img_scaled)
    plt.axis('off')
    plt.title(f'Compressed Image (K={k})')
    plt.show()



```

```python
for k in [16, 32, 64]:
    compress_with_kmeans(k)
```
